```python
housing_extended['tradeTime'] = pd.to_datetime(housing_extended['tradeTime'])
```

```python
fig1 , ax1 = plt.subplots(figsize=(8,8))
df2=housing_extended[ housing_extended.tradeTime > '2010-01-01']
df1=df2.resample('M' ,on='tradeTime').mean()
df3=df1.iloc[:,14]
df3.plot(ax=ax1 ,color='crimson')
ax1.set_ylabel('pricePerSquare')
ax1.set_title('pricePerSquare over time');
```

```python

fig2 , ax2 =  plt.subplots(figsize=(8,8))
df4=housing_extended[ housing_extended.tradeTime > '2009-01-01']
df5=df4[df4['distanceToCapital']<15]
df6=df5.resample('3M' ,on='tradeTime').size()
df7=df6[df6.index > '2009-10-31']

df7.plot(ax=ax2 ,color='orange')
#df7=df4.sort_values(by='tradeTime')

ax2.set_ylabel('frequency')
ax2.set_title('center city house trades over time');
```
